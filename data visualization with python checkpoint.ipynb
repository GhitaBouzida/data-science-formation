```python
# Extract Japanese sales and Other sales columns
japanese_sales = data['JP_Sales']
other_sales = data['Other_Sales']

# Creating the scatter plot
plt.scatter(japanese_sales, other_sales)
plt.xlabel('JP_Sales (in millions)')
plt.ylabel('Other_Sales (in millions)')
plt.title('Relationship between Japanese and Other Sales')

# Display the plot
plt.show()
```

```python
# Extract platform sales columns
platforms = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
platform_sales = data[platforms].T

# Creating the stacked area chart
plt.stackplot(data['Year'], platform_sales, labels=platforms)
plt.xlabel('Years')
plt.ylabel('Sales (in millions)')
plt.title('Distribution of Sales by Platform over the Years')
plt.legend(loc='upper left')

# Display the plot
plt.show()
```

```python
# Group data by publisher and calculate global sales
publisher_sales = data.groupby('Publisher')['Global_Sales'].sum().nlargest(10)

# Creating the bar chart
plt.bar(publisher_sales.index, publisher_sales.values)
plt.xlabel('Publishers')
plt.ylabel('Global Sales (in millions)')
plt.title('Top 10 Publishers by Global Sales')

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Group data by genre and calculate global sales
genre_sales = data.groupby('Genre')['Global_Sales'].sum()

# Creating the pie chart
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribution of Genres by Global Sales')

# Display the plot
plt.show()
```
